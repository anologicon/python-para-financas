```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('./Housing.xlsx')
```

```python
data
```

```python
X = data[['House Size (sq.ft.)', 'Number of Rooms', 'Year of Construction']]
Y = data['House Price']
```

```python
X1 = sm.add_constant(X)
reg = sm.OLS(Y, X).fit()

reg.summary()
```
