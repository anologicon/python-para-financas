```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

 - ^GSPC  = S&P500
 - ^IXIC  = NASDAQ
 - ^GDAXI = DAX
 - ^BVSP  = IBOVESPA

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI','^BVSP']

ind_data = pd.DataFrame()

for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='1997-1-1')['Adj Close']
```

```python
ind_data.head()
```

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize = (15,6))
plt.show()
```

Retorno do indice

```python
ind_retorno = (ind_data / ind_data.shift(1)) - 1

ind_retorno.tail()
```

```python
ind_retorno_anual = ind_retorno.mean() * 250
ind_retorno_anual
```
