# Prevendo lucro bruto futuro

* Requisitos:

- Receita esperada

- Custo da Mercadoria Vendida (CMV)

```python
import numpy as np
import matplotlib.pyplot as plt
```

Receitas esperadas deste ano é de 170 mi
Desvio padrão é de 20 mi

```python
rev_m = 170
rev_stdev = 20
interations = 2000
```

Gerando 1000 valores aleatorios, apartir do rev_m que é a receita esperada, e variando com o desvio padrão rev_stdev

```python
receita = np.random.normal(rev_m, rev_stdev, interations)
```

Plotando a distribuição dos dados

```python
plt.figure(figsize=(15,6))
plt.plot(receita)
plt.show()
```

Pensando que o custo é 60% da receita mas isto é um exemplo

E a estimativa do desvio padrão do cmv é de 10%

```python
custo = 0.6
std = 0.1


CMV = - (receita * np.random.normal(custo,std))

plt.figure(figsize=(15,6))
plt.plot(CMV)
plt.show()
```

```python
CMV.mean()
```

```python
CMV.std()
```

```python
lucro_bruto = receita + CMV
```

```python
plt.figure(figsize=(15,6))
plt.plot(lucro_bruto)
plt.show()
```

Maior valor potencial do lucro bruto

```python
max(lucro_bruto)
```

Menor valor potencial do lucro bruto

```python
min(lucro_bruto)
```

```python
lucro_bruto.mean()
```

```python
lucro_bruto.std()
```

```python
plt.figure(figsize=(15,6))
plt.hist(lucro_bruto, bins = [40,50,60,70,80,90,100,110,120])
plt.show()
```
