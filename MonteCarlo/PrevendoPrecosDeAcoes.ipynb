# Precificando os papeis de uma empresa

## Preço de Hoje = Preço de Ontem * e ^ r


**r** : retorno logaritimico entre o preço da ação de ontem e de hoje

r  = logaritimo(preço de hoje / preço de ontem)

## Movimento Browniano

É utilizado para modelar o r:

### Drift:

A direção que as taxas de retorno tiveram no passado = ln(preço de hoje / preço de ontem)

Podemos utilizar o retorno diario da ação no passado

### Volatilidade:

Variavel aleatoria = volatilidade historica * Z(Rand(0;1)) ou um numero aleatorio entre 0 e 1

O numero entre 0 e 1 é um percentual

### No final


## Preço hoje = Preço ontem * e ^ (Drift + Volatilidade)

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
ticker = 'KLBN4.SA'

data = pd.DataFrame()
data[ticker] = wb.DataReader(ticker, data_source='yahoo', start='2007-1-1')['Adj Close']
```

```python
retorno_log = np.log(1 + data.pct_change())
```

```python
retorno_log.tail()
```

```python
data.plot(figsize=(10,6))
```

```python
retorno_log.plot(figsize=(10,6))
```

```python
media = retorno_log.mean()
```

```python
variancia = retorno_log.var()
```

## Drift

```python
drift = media - (0.5 * variancia)
drift
```

# Volatilidade

```python
std = retorno_log.std()
std
```

O Z 

Se o evento tem 95% de chance de acontecer, as distancia do evento e a media seria de 1.64 desvios padrão

```python
norm.ppf(0.95)
```

Gerando números aleatorios

```python
x = np.random.rand(10,2)
x
```

Calculando o desvio padrão das distancias de cada valor gerando aleatoria mente

X é uam matriz de probabiliaddes

```python
norm.ppf(x)
```

```python
Z = norm.ppf(np.random.rand(10,2))
Z
```

```python
intervalos = 2 # prever o preço para os proximos 1000 dias
interacoes = 10 # 10 series de previsões futuras
```

```python
retorno_diario = np.exp(drift.values + std.values * (norm.ppf(np.random.rand(intervalos, interacoes))))
```

```python
retorno_diario
```

Agora precisamos criar um tipo de iteração para prever os próximos preços


preçoHoje = PreçoOntem * retorno_diario

Si+999 = St+9998 * Rt+999

```python
S0 = data.iloc[-1]
S0

# Ultimo valor da lista ou o ultimo preço
```

```python
lista_precos = np.zeros_like(retorno_diario)

# Cria uma matriz com a mesma estrutura da matriz retorno_diario porem com os valores zerados
```

```python
lista_precos[0] = S0
lista_precos

# Setando o primeiro valor para começar o loop de calculo dos proximos valores
```

Os valores da matriz serão preenchidos com os valores da matriz de retornos diarios


**Atenção!**

Essa técnica faz com que poçamos utilizar o valore previsto de ontem apra prever o próximo

```python
for t in range(1, intervalos):
    lista_precos[t] = lista_precos[t - 1] * retorno_diario[t]
```

```python
lista_precos
```

```python
plt.figure(figsize=(10,6))
plt.plot(lista_precos);
```
